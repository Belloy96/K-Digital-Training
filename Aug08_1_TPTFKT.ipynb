```python
sentence = """도미노피자의 로스트아크 굿즈 스페셜 딜 프로모션은 9일부터 31일까지 온라인을 통해 프리미엄 피자(L) 주문 시 로스트아크 캐릭터를 활용한 키보드 교체형 버튼인 '키캡'과 '콜라 1.25ℓ'로 구성된 굿즈 스페셜 세트를 합리적인 가격에 구매할 수 있다. 이번 로스트아크 굿즈 스페셜 딜 프로모션의 굿즈인 키캡은 로스트 아크의 인기 캐릭터인 모코코와 에스더 총 2종으로 구성되어 있어 로스트아크 팬들의 큰 관심을 끌 것으로 기대된다. 더불어 도미노피자는 로스트아크 굿즈 스페셜 딜 프로모션과 함께 로스트아크 테마가 적용된 피자박스도 함께 선보일 예정이다. 이 피자 박스에는 로스트아크의 여러 캐릭터가 그려져 있어 로스트아크를 즐기는 팬들에게 더 큰 즐거움을 줄 것으로 보인다. 도미노피자 관계자는 "31일까지 맛있는 도미노피자도 즐기고 귀여운 로스트아크의 캐릭터들도 만나볼 수 있는 기회를 놓치지 마시길 바란다"고 말했다."""
sentence
```

```python
sentence2 = sentence.replace(" ", "") # 띄어쓰기 없애기
sentence2
```

```python
from pykospacing import Spacing
```

```python
spacing = Spacing()
s = spacing(sentence2)

print(sentence)
print()
print(s)
```

```python
# SNS나 채팅의 경우 'ㅋㅋㅋㅋㅋㅋㅋㅋㅋㅋ', 'ㅎㅎㅎㅎㅎㅎㅎㅎㅎㅎ' 같이
# 불필요하게 연속되는 경우가 꽤나 많은데
# 반복되는 것들을 하나로 정규화시켜줄 수 있음
print(emoticon_normalize('ㅋㅋㅋㅋㅋㅋㅋㅋㅋㅋ앜ㅋㅋㅋㅋㅋㅋㅋㅋㅋ', num_repeats=2))
print(emoticon_normalize('ㅎㅎㅎㅎㅎㅎㅎㅎㅎㅎ앜ㅋㅋㅋㅋㅋㅋㅋㅋㅋ', num_repeats=2))
print(emoticon_normalize('ㅠㅠㅠㅠㅠㅠㅠㅠㅠㅠ앜ㅋㅋㅋㅋㅋㅋㅋㅋㅋ', num_repeats=2))
print(emoticon_normalize('ㅜㅜㅜ앜ㅋㅋㅋㅋㅋㅋㅋㅋㅋ', num_repeats=2))
```
