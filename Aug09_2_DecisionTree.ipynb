```python
# target 실제값을 수치로 변환한 값 출력
# 0 : 양성 / 1 : 음성
print(cancer.target)
```

```python
# 이 데이터를 훈련용, 테스트용으로 나누고, 시험지, 정답지로 나누기
x = cancer.data
y = cancer.target

x_train, x_test, y_train, y_test = train_test_split(x, y, test_size = 0.2, random_state = 1024, stratify = cancer.target)

# 구성비가 다른 불균형 상황에서는 stratify 옵션 넣어주기
# train, test 데이터에 있는 label 값의 비율이 서로 다르면
# 학습 결과나 모델 평가 등에서 잘 못된 영향을 끼칠 가능성이 있음
# dataset 쪽의 클래스의 비율과 같은 비율로 조정

# 내가 쓴 답
# train_data, test_data, status_train, status_test = train_test_split(data, cancer.target, test_size = 0.3)
```

```python
# 의사결정나무 모델 생성 (깊이 = 4, 불순도 = 지니계수)
cancer_tree = DecisionTreeClassifier(max_depth = 4, criterion = 'gini') # 'gini'가 기본값이기 때문에 굳이 명시를 하지 않아도 됨
```

```python
# 모델 학습 (분리하기 전 데이터)
cancer_tree.fit(x, y)
```

```python
# dot => png로 나타내는 작업
# 열의 이름으로 분류하게 => 예측할 클래스의 이름을 '양성', '음성'
import graphviz
from sklearn.tree import export_graphviz
from subprocess import check_call

cancerTree = export_graphviz(
    cancer_tree,
    out_file = None,
    feature_names = cancer.feature_names,
    class_names = ['양성', '음성'],
    filled = True,
    rounded = True,
)

graph = graphviz.Source(cancerTree)
graph
```
